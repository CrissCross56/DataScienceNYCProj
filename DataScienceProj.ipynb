```python
borough_table = pd.pivot_table(
    data = df,
    index = 'borough',
    aggfunc = 'mean',
    values = ['number_of_persons_injured' , 'number_of_persons_killed']
)
```

```python
df['injured_people'] = (df["number_of_persons_injured"] > 0).astype(int)
df['dead_people'] = (df["number_of_persons_killed"] > 0).astype(int)
```

# Injuries and Deaths per Borough

```python
borough_df = (
    df.groupby("borough")[["injured_people", "dead_people"]]
      .sum()
)

borough_df.plot(kind="bar", width=0.7, logy=True)
```
