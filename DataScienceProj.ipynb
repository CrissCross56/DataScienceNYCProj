```python
df = df.drop(columns=[':version', ':created_at',':updated_at',])
```

```python
df.dtypes
```

```python
df["crash_date"] = pd.to_datetime(df["crash_date"])
```

```python
df['crash_time'] = pd.to_datetime(df['crash_time'], format="%H:%M")
```

df.pivot_table(index='Row_Column', columns='Column_Header', values='Data_Column', aggfunc='mean')


```python
collision_table = pd.pivot_table(
    data = df,
    index = 'contributing_factor_vehicle_1',
    aggfunc = 'mean',
    values = ['number_of_persons_injured', 'number_of_persons_killed']
).sort_values('number_of_persons_killed',ascending=False)
```

```python
df['injuries'] = (df['number_of_persons_injured']>0)
df['deaths'] = (df['number_of_persons_killed']>0)

df['injuries']
```

```python
df['crash_date'] = pd.to_datetime(df['crash_date'])
```

```python
caused_death = df[ df['deaths'] == True ]
no_deaths = df[ df['deaths'] == False ]
```
